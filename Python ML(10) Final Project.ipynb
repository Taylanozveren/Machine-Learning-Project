```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv"
df = pd.read_csv(filepath)
```

## Data Preprocessing

### One Hot Encoding


#### In this step, the pd.get_dummies function is used to convert categorical variables into numerical values. One-Hot Encoding is a method of converting categorical variables into numerical values. 
#### It creates a separate column for each unique category, and if an observation belongs to that category, a value of 1 is assigned to that column; otherwise, a 0 is assigned. This method transforms categorical data into a format that models can understand. For example, in the WindGustDir column, if there are directions like "N", "S", "E", "W", a separate column is created for each direction, and if the observation belongs to a particular direction, it gets a 1 in that column, and 0 in the others.

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

#### Next, we replace the values of the 'RainTomorrow' column changing them from a categorical column to a binary column. We do not use the `get_dummies` method because we would end up with two columns for 'RainTomorrow' and we do not want, since 'RainTomorrow' is our target.


```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

## Training Data and Test Data


```python
df_sydney_processed.drop('Date',axis=1,inplace=True)  #Since time-related information like dates doesn't have a direct meaning for model training, this column has been excluded.
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

### Linear Regression


### 1. Step: Splitting the dataset into training and test sets
#### We will use the train_test_split function from Python to split the dataset into training (80%) and test (20%) sets. The following code achieves this:

```python
from sklearn.model_selection import train_test_split

# Defining the features (X) and target (y) columns
X = df_sydney_processed.drop(columns=['RainTomorrow'])
y = df_sydney_processed['RainTomorrow']

# Splitting the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

### 2. Step: Creating and training a Linear Regression model
#### In this step, we will create a LinearRegression model and train it using x_train and y_train.

```python
from sklearn.linear_model import LinearRegression

# Creating the Linear Regression model
LinearReg = LinearRegression()

# Training the model with the training data
LinearReg.fit(x_train, y_train)
```

### 3. Step: Model Predictions
#### The model (LinearReg) uses the test data (x_test) to make predictions based on what it has learned during training. The predicted values are stored in the variable predictions.

```python
predictions = LinearReg.predict(x_test)
```

### 4.Step: Evaluating Model
### Model Performance Summary

- **MAE (Mean Absolute Error - 0.2563)**: Indicates how much the model's predictions deviate from the actual values on average. Lower is better.

- **MSE (Mean Squared Error - 0.1157)**: The average of squared errors, penalizing larger errors more. Lower is better.

- **R² (R-Squared - 0.4271)**: Shows how much of the variation in the target variable is explained by the model. The closer to 1, the better. In this case, it explains 42.71% of the variance.

In summary, the error values are reasonable, but R² should ideally be higher. The model can be improved.


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculating MAE, MSE, and R^2 values
LinearRegression_MAE = mean_absolute_error(y_test, predictions)
LinearRegression_MSE = mean_squared_error(y_test, predictions)
LinearRegression_R2 = r2_score(y_test, predictions)

print("MAE:", LinearRegression_MAE)
print("MSE:", LinearRegression_MSE)
print("R^2:", LinearRegression_R2)
```

### 5.Step: Show the MAE, MSE, and R2 in a tabular format using data frame for the linear model.

```python
# Creating a DataFrame to display the results in tabular format
metrics_df = pd.DataFrame({
    'Metric': ['MAE', 'MSE', 'R2'],
    'Value': [LinearRegression_MAE, LinearRegression_MSE , LinearRegression_R2]
})

# Displaying the DataFrame
metrics_df
```

## KNN
### 6.Step: Create and train a KNN model called KNN using the training data (x_train, y_train) with the n_neighbors parameter set to 4.

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=4)
KNN.fit(x_train, y_train)
```

### 7.Step: Now use the predict method on the testing data (x_test) and save it to the array predictions.7.Step: Now use the predict method on the testing data (x_test) and save it to the array predictions.


```python
predictions = KNN.predict(x_test)
```

### 8.Step: Using the predictions and the y_test dataframe calculate the value for each metric using the appropriate function.

```python
from sklearn.metrics import accuracy_score, jaccard_score, f1_score

# Accuracy: The ratio of correctly predicted instances to the total instances. It shows the overall correctness of the model.
KNN_Accuracy_Score = accuracy_score(y_test, predictions)

# Jaccard Index: Measures the similarity between predicted positive cases and actual positive cases. It calculates the intersection over the union of positive predictions.

KNN_JaccardIndex = jaccard_score(y_test, predictions)

# F1 Score: The harmonic mean of precision and recall. It balances the trade-off between precision and recall for positive class predictions.
KNN_F1_Score = f1_score(y_test, predictions)


print(f"KNN_Accuracy_Score: {KNN_Accuracy_Score:.3f}")
print(f"KNN_JaccardIndex: {KNN_JaccardIndex:.3f}")
print(f"KNN_F1_Score: {KNN_F1_Score:.3f}")
```
